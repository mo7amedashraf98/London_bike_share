### Data Cleaning

```python
# Renaming the columns

new_col = {'timestamp': 'time',
           'cnt': 'count',
           't1': 'temp_real',
           't2': 'temp_feels_like',
           'hum': 'humidity_percent',
           'wind_speed': 'wind_speed_kph',
           'weather_code': 'weather',
           'is_holiday': 'holiday',
           'is_weekend': 'weekend'}

df.rename(new_col, axis=1, inplace=True)

df.head()
```

```python
# Changing the weather codes to names
weather_name = {1.0: 'Clear',
                2.0: 'Scattered Clouds',
                3.0: 'Broken Clouds',
                4.0: 'Cloudy',
                7.0: 'Rain',
                10.0: 'Rain With Thunderstorms',
                26.0: 'Snowfall'}
# Maping the newly created dictionary and changing the column type
df['weather'] = df['weather'].map(weather_name).astype(str)
```

```python
# Changing the season codes to names
season_name = {0.0: 'Spring',
          1.0: 'Summer',
          2.0: 'Autummn',
          3.0: 'Winter'}
# Maping the newly created dictionary and changing the column type
df['season'] = df['season'].map(season_name).astype(str)
```
